```python
##this is a program for checking presence of hydrocarbon at a depth
#importing the libraries
import numpy as np
import pandas as pd
```

```python
#reading the data
data=pd.read_csv("C:\\Users\\admin\\Documents\\logging.csv")
```

```python
data
```

```python
#decision tree classification
x=data.iloc[:,:-1]
features=(['SP log','GR','Caliper','Resistivity','Neutron porosity density crossover'])
y=data.iloc[:,5].values
from sklearn.preprocessing import LabelEncoder
lblen_x=LabelEncoder()
x=x.apply(LabelEncoder().fit_transform)
from sklearn.tree import DecisionTreeClassifier
regressor=DecisionTreeClassifier(criterion='gini',max_depth=None,splitter='best')
print(x,features)
```

```python
#training the data
regressor.fit(x,y)
```

```python
#prediction
x_in=np.array([0,0,0,0,0])
x_in
y_pred=regressor.predict([x_in])
print("presence of hydrocarbon at this depth=",y_pred)
```

```python
# accuracy
regressor.score(x,y)*100
```
